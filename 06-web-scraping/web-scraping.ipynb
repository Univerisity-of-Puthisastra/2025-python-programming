```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "MyWebScraperBot/1.0 (contact@example.com)"
}

response = requests.get("https://en.wikipedia.org/wiki/Web_scraping", headers=headers)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
p_tags = soup.find(id='mw-content-text').find_all('p')

for p in p_tags:
    print(p.get_text())
```

```python
from selenium import webdriver

driver = webdriver.Chrome()
```

```python
driver.get("https://www.google.com")
```

```python
from IPython.display import Image, display

display(Image(driver.get_screenshot_as_png()))
```

```python
from selenium.webdriver.common.by import By
text_area = driver.find_element(By.TAG_NAME, "textarea")
text_area.send_keys("Hello, World!")
```

```python
from selenium.webdriver.common.keys import Keys

text_area.send_keys(Keys.ENTER)
```

```python
driver.quit()
```
