```python
import requests

response = requests.get("http://localhost:8000/api/v1/test")
print(response.json())
```

```python
response = requests.get("http://localhost:8000/api/v1/test", auth=("admin", "password"))
print(response.text)
```

```python
response = requests.get("http://localhost:8000/api/v1/test/jwt", auth=("admin", "password"))
print(response.text)
```

```python
credential_response = requests.post("http://localhost:8000/api/v1/login", headers={"Content-Type": "application/json"}, data='{"username": "admin", "password": "password"}')
credential_response.json()
```

```python
credential_response = requests.post("http://localhost:8000/api/v1/login", json={"username": "admin", "password": "password"})
credential_response.json()
```

```python
response = requests.get("http://localhost:8000/api/v1/test/jwt", headers={"Authorization": f"Bearer {credential_response.json()['token']}"})
print(response.text)
```
